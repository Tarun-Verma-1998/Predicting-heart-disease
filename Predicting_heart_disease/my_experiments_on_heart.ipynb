```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# To check number of people of different ages
sns.distplot(df['age'])
```

```python
#to calculate distinct values in chest pain or types of chest pains
set(df['cp'])
```

```python
x = df['cp'].value_counts()
x
```

```python
# Pie chart to check which type of chest pain is more in patients

fig,ax = plt.subplots(figsize=(7,7))

labels = list(x.index)
sizes = list(x.values)

plt.pie(sizes,labels=labels,shadow=True,autopct='%1.1f%%')
plt.show()
```

```python
# FacetGrid to check in every type of chest pain what is the cholestrol
x = sns.FacetGrid(row='cp',data=df)
x.map(plt.hist,'chol',edgecolor='black',color='green') #sharex
```

```python
sns.barplot(x='sex',y='chol',hue='target',data=df)
```

```python
set(df['restecg'])
```

```python
# how many people are suffering from heart disease

fig,ax = plt.subplots(1,2,figsize=(14,7))
sns.countplot(x='target',data=df,ax=ax[0])
ax[0].set_xlabel('Disease Count \n [0]-->No [1]-->Yes')
ax[0].set_title("Heart Disease")

df['target'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,cmap='Greens',explode=[0.1,0.0])
```

# FEATURE ANALYSIS

```python
# a-->  sex (category)

fig,ax = plt.subplots(1,2,figsize=(14,7))

sns.countplot(x='sex',data=df,ax=ax[0],hue='target')
ax[0].set_xlabel('[0]-->Female  [1]-->Male')

df['sex'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,cmap='Greens',explode=[0.1,0.0],ax=ax[1])
ax[1].set_title('[0]-->Female  [1]-->Male')
```

```python
# fasting blood sugar (fbs)(category)

fig,ax = plt.subplots(1,2,figsize=(14,6))

sns.countplot(x='fbs',hue='target',data=df,ax=ax[0])

df['fbs'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,cmap='Greens',explode=[0.1,0.0],ax=ax[1])
ax[1].set_title('[0]--> fbs<120   [1]--> fbs>120')
```

```python
# restecg (0,1,2) (category)
fig,ax = plt.subplots(1,2,figsize=(14,6))

sns.countplot(x='restecg',hue='target',data=df,ax=ax[0])
ax[0].set_xlabel('resting electrocardiographic')

df['restecg'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,cmap='Greens',explode=[0.0005,0.005,0.05],ax=ax[1])
ax[1].set_title('resting electrocardiographic')
```
